# GROUPBY ->Group the custommer by name and sum their data.

```python
abc=data.groupby('name')
total=abc.sum()
total
```

```python
xyz=customer.groupby('name')
total=xyz.sum()#sum(),mean(),max()
total
```

```python
#filter
total.filter(regex='^K',axis=0)
```

```python
total.T
```

```python
#filtering using re!
list(enumerate(total.index))
```

```python
l=[5,9,4]
print(list(enumerate(l)))
```

```python
import re
total.loc[[values for index,values in enumerate(total.index) if re.search(r'H[a-z]+',values)]]
```

```python
import matplotlib.pyplot as plt
bar_plot=total.plot(kind='bar',legend=None, title='total sales',figsize=(16,7))
bar_plot.set_xlabel("customer")
bar_plot.set_ylabel("sales($)")
plt.show()
```
