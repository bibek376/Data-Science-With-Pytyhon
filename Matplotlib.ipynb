```python
plt.figure(figsize=(10,6))
plt.plot(age,salary,color='red',linestyle='--',label='salary by age')
plt.plot(age,exp,color='k',linestyle=':',label='salary by exp')
plt.legend()
plt.grid(True)
plt.title('age vs salary/exp')#for main title
plt.xlabel('ages')# for  x-axis defination
plt.ylabel('salary expenses')# for y-axis defination
plt.savefig('abc.png')
plt.tight_layout()
#like strip() function
plt.show()
```

# Shadow() and startangle() function

```python
#shadow is used for dim boaarder 
#startange is used for initial angle
marks=[45,78,88,79,94]
subject=['physic','nepali','math','chemistry','english']
plt.pie(marks,labels=subject,shadow=True,startangle=60)
plt.title("Carlos Marks:")
plt.show()
```

# explode(),autopct(),wedgeprops()and textprops() functions

```python
#explode--->explode/separate pie components
#aautopct--->% value on pie components
#wedgeprops--->adding edgecolor
#textprops----->text color
```

```python
plt.figure(figsize=(10,10))
marks=marks=[45,78,88,79,94]
subject=['physic','nepali','math','chemistry','english']
explode=[0.1,0,0.2,0,0]
plt.pie(marks,labels=subject,explode=explode,shadow=True,startangle=70,
       autopct="%0.0f%%",wedgeprops={'edgecolor':'k'},textprops=dict(color='k'),colors=['r','b','g','pink','y'])
plt.title("Carlos Marks")
plt.show()
```

# width() function

```python
import matplotlib.pyplot as plt
import numpy as np
before=np.array([10,11,9,12,4])
now=np.array([3,10,7,11,12])
after=np.array([11,12,8,17,10])
plt.bar([1,3,5,7,9],before,color='red',label='before',width=0.4)
plt.bar([2,4,6,8,10],after,color='blue',label='after',width=1)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
xloc=np.arange(len(before))
plt.bar(xloc+0.3,before,width=0.3,color="red",label="before")
plt.bar(xloc,after,width=0.3,color='blue',label='after')
plt.legend()
plt.title("Bar Example")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(15,6))
xloc=np.arange(len(before))
width=0.2
labelsx=['Group'+x for x in list('ABCDE')]
labelsy=list(range(after.max()))
plt.bar(xloc+width,before,width,color='green',label='before')
plt.bar(xloc,after,width,color='red',label='after')
plt.bar(xloc-width,now,width,color='blue',label='now')
plt.legend()
plt.title("Graph Example")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(10,6))#ax--->one part of main figure
xloc=np.arange(len(before))
width=0.25
ax.bar(xloc+width,before,width,color='g',label='Before')
ax.bar(xloc,after,width,color='r',label='after')
ax.bar(xloc-width,now,width,color='b',label='now')
ax.legend(loc='upper right')
ax.yaxis.grid(True)
#ax.xaxis.grid(True)
ax.set_xticks(ticks=range(len(before)))#for x-axis value
#ax.set_xticks(ticks=['a','b','c','d','e'])#not working
#ax.set_xticklabels(before)
#ax.set_xticklabels(['a','b','c','d','e'])
ax.set_yticks(ticks=range(0,21))#for y-axis value
#ax.set_yticklabels(after)
plt.title("Advance Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(9,6))
xloc=np.arange(len(before))
width=0.25
labelsx=['Group'+x for x in list("ABCDEFG")]
labelsy=list(range(after.max()))
ax.barh(xloc+width,before,width,color='b',label=before)
ax.barh(xloc,after,width,color='r',label=after)
ax.barh(xloc-width,now,width,color='g',label=now)
ax.yaxis.grid(True)
ax.set_yticks(ticks=range(len(before)))
#ax.set_yticklabels(before)
ax.set_xticks(ticks=range(0,21))
plt.title("Advance Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
